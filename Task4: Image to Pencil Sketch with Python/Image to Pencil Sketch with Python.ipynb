### Task4: Image to Pencil Sketch with Python

```python
#Import libraries
import numpy as np
import cv2
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
#Get image location
img_file = "puppy.jpg"

#readin the image
img = cv2.imread(img_file)
```

```python
def show_image(img):
    cv2.imshow('Image' , img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
show_image(img)
```

```python
#Display Image
plt.figure(num = 1, figsize=(5,5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize=15)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
#convert the image to grey color
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show_image(gray_image)
```

```python
#Display Image
plt.figure(num = 1, figsize=(5,5))
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.title('Gray Image', fontsize=15)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
#invert the image

inverted_gray_image = 255-gray_image
show_image(inverted_gray_image)
```

```python
#Blur the image using dodging function
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21,21),0)
show_image(blurred_image)
```

```python
#Display Image
plt.figure(num = 1, figsize=(5,5))
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image', fontsize=15)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# invert the blurred image

inverted_blurred_image = 255-blurred_image
show_image(inverted_blurred_image)
```

```python
#Display Image
plt.figure(num = 1, figsize=(5,5))
plt.imshow(cv2.cvtColor(inverted_blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image', fontsize=15)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# create the pencil sketch image
pencil_sketch_img = cv2.divide(gray_image,inverted_blurred_image, scale=256.0)
show_image(pencil_sketch_img)
```

```python
 #Display Image
plt.figure(num = 1, figsize=(5,5))
plt.imshow(cv2.cvtColor(pencil_sketch_img, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image', fontsize=15)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
cv2.imwrite('output_img.jpg', (pencil_sketch_img))
```
